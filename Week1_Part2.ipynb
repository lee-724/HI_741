### Dictionaries

Dictionaries are not sequences like string and list; they are mappings. There is no order in dictionaries. 

```python
d = {'name': 'Essential programming for health informatics',
     'code': 'HI 741',
     'status': 'required',
     'enrolled': 30}
```

```python
### Key-value pair
```

```python
d['name']
```

```python
d['enrolled']
```

```python
d
```

```python
d['enrolled'] += 1
d
```

```python
# ways to create dictionaries
# 1
d1 = {}
d1['name'] = 'Essential programming for health informatics'
d1['code'] = 'HI 741'
d1['status'] = 'required'
d1['enrolled'] = 30
```

```python
d1
```

### Tuples

Tuples are sequences but are immutable. Ideal for cases when you know the length is fixed and will no longer be changed.

```python
t = (1,2,3)
l = [1,2,3]
```

```python
t[0]
```

```python
t[1]
```

```python
l
```

```python
l[0] = -1
l
```

```python
t[0] = -1
```

### Set

Sets are unordered collections of unique and immutable objects. 
Sets are not sequences.

```python
a = {0,1}
b = {1,0}
```

```python
x = {'a','b','c','d'}
y = {'a','z','e'}
```

```python
z = {0,0}
```

```python
z
```

```python
# Set operations
x & y # intersection
```

```python
x | y # union
```

```python
x - y # difference
```

```python
x > y # superset
```
