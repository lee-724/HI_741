### Pythonic programming

#### Enumerate 

```python
browsers = ['Chrome','Firefox','Opera','Vivaldi']
# create an enumeratable and convert to list
x = list(enumerate(browsers))
print(x)
print (dict(enumerate(browsers)))
print (set(enumerate(browsers)))
```

```python
browsers = ['Chrome','Firefox','Opera','Vivaldi']
# create an enumeratable and convert to list
x = list(enumerate(browsers, start = 1))
print(x)
```

#### Zip

```python
# The zip() function returns a zip object, 
# which is an iterator of tuples where the first item in each passed iterator is paired together, 
# and then the second item in each passed iterator are paired together etc.
languages = ['Java', 'Python', 'JavaScript']
versions = [14, 3, 6]

result = zip(languages, versions)

print(list(result))
```

#### Collections 
This module implements specialized container datatypes providing alternatives to Python’s general purpose built-in containers, dict, list, set, and tuple.


```python
# Counter  
from collections import Counter
print (Counter('abcdeabc'))
```

```python
counter = Counter('abcdeded')
print (counter['d'])
list(counter.elements())
```

```python
counter.most_common(2)
```

```python
counter.most_common(1)
```

```python
# defaultdict
# defaultdict never raises a KeyError. It provides a default value for the key that does not exists.

from collections import defaultdict
sentence = "I feel good today I feel good"
words = sentence.split(' ')
print(words)

d = defaultdict(int)
for word in words:
    d[word] += 1

print(d)
```

```python
# ordereddict
import collections
#Create normal dict
# In Python versions 3.7 and later, the standard dict preserves the insertion order of key-value pairs. This means that when you iterate over a dictionary, the items are returned in the order they were added.
d = {}
d['A'] = 1
d['B'] = 2
d['C'] = 3
d["E"] = 5
d['D'] = 4
for item in d.items():
    print(item)
print()
#Create ordered dict
d = collections.OrderedDict()
d['A'] = 1
d['B'] = 2
d['C'] = 3
d['D'] = 4
for item in d.items():
    print(item)
```

```python
# Ordered dict supports ordered-based operations
# move_to_end(key, last=True): Moves the specified key to either the beginning or end of the dictionary.
d.move_to_end('A', last=True)  # Moves 'apple' to the beginning
```

```python
d
```

Standard dictionaries tend to be more memory-efficient and faster because they don't need to maintain extra data structures for the order of insertion.

#### List comprehension

```python
l = [1, 2,3]
l2 = []
for i in l:
    new_i = i * 2
    l2.append(new_i)
l2
```

```python
l = [1, 2, 3]
l2 = [x * 2 for x in l]
l2
```
