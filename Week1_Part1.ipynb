### Strings 

#### String is a sequence. It has order, from left to right.
#### Strings are immutable.

#### Official Python resource for strings: https://docs.python.org/3/library/stdtypes.html#textseq

```python
a = 'Python'
len(a)
```

```python
# Indexing
a[0]
```

```python
# Backward indexing
a[-1]
```

```python
a[-2]
```

```python
# Slicing 
a[0:2]
```

```python
# More slicing
print (a[1:]) # everything but the first element
print (a[:5]) # everything but the last element
print (a[:len(a)-1]) # everything but the last element
print (a[:-1]) # everything but the last element
```

```python
# Concatenation 
b = 'Program'
a + b
```

```python
a + ' ' + b
```

```python
# Repeatition
a * 10
```

```python
# Strings are immutable: the original copy is not changed by any operations.
a
```

```python
# Each operation will produce a new string 
c = a + b
c
```

```python
# If you try to change an immutable object...
a[0] = 'Q'
a
```

```python
# Some special methods for string
a.find('b') # Find the offset of a substring
# -1 means it does not exist in the string
```
