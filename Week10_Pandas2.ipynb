#### Creating Pandas dataframe

```python
data = {'col_1': [3, 2, 1, 0], 'col_2': ['a', 'b', 'c', 'd']}
df = pd.DataFrame.from_dict(data)
```

#### Retrieving data from data frame

```python
df = pd.DataFrame([[1, 2], [4, 5], [7, 8]],
    index=['cobra', 'viper', 'sidewinder'],
    columns=['max_speed', 'shield'])
```

```python
df.head()
```

```python
df.loc["cobra", "shield"]
```

```python
df.loc["viper"]
```

#### Modifying values in data frame

```python
df.loc['cobra'] = 10
```

```python
df
```

```python
df.loc[['viper', 'sidewinder'], ['shield']] = 50
df
```

```python
# Modifying cell values based on conditions 
df.loc[(df['shield'] > 35) & (df["max_speed"] < 10)] = 0
df
```

#### Iterating through a data frame

```python
# Iterate over DataFrame rows as (index, Series) pairs.
for idx, row in df.iterrows():
    print(idx)
    print(row)
```

#### Removing data from a data frame

```python
df = pd.DataFrame([('falcon', 'bird', 389.0),
                    ('parrot', 'bird', 24.0),
                    ('lion', 'mammal', 80.5),
                    ('monkey', 'mammal', np.nan)],
                  columns=('name', 'class', 'max_speed'))
df
```

```python
df.pop('class')
df
```
