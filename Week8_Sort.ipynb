```python
def bubbleSort(array):
    for i in range(len(array)):
        for j in range(0, len(array) - i - 1):
#             print(i, j)
            if array[j] > array[j + 1]:
                temp = array[j]
                array[j] = array[j+1]
                array[j+1] = temp
    return array
    
data = [-1, 4, 0, 5, -3]
bubbleSort(data)
```

```python
def insertionSort(lst):
    if len(lst) <= 1:
        return  
    # Start from the second element in the list 
    for i in range(1, len(lst)):  
        key = lst[i]  
        j = i-1
        while j >= 0 and key < lst[j]:  
            lst[j+1] = lst[j]  
            j -= 1
        lst[j+1] = key  
        print("Pass: " + str(i), "Key element is: " + str(lst[i]), lst)
 
lst = [-1, -4, 0, 10, -10]
insertionSort(lst)
print(lst)
```
