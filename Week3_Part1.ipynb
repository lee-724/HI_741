```python
# Iteration 
l = [0, 1, 2, 3]
for e in l:
    print (e)
```

```python
# While loop
i = 0
while i < len(l):
    print(l[i])
    i += 1
```

```python
# Loop exit
# What does this loop do?
i = 0
while i < len(l):
    if l[i] % 2 != 0:
        print (l[i])
        break
    else:
        i += 1
```

```python
# Loop continue
i = 0 
while i < len(l):
    if l[i] % 2 == 0:
        i += 1
        continue
    else:
        print(l[i])
        i += 1
```

#### General rules for Python statements

1. Parentheses are optional

```python
i = 0 
while i < len(l):
    if (l[i] % 2 == 0):
        i += 1
        continue
    else:
        print(l[i])
        i += 1
```
