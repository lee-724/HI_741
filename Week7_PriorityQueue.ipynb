Combine your Patient and PriorityQueue class to simulate a real patient queue waiting to be seen in a hospital. 
Now assume that each patient will be assigned a severity score, and the order to be seen should be based on the severity score, e.g., higher scores need to be seen before lower scores.

```python
class Patient:
    def __init__(self, id, priority):
        self.id = id
        self.priority = priority
```

```python
p1 = Patient(0, 2)
p2 = Patient(1, 1)
p3 = Patient(2, 5)
p4 = Patient(3, 0)
```

```python
class PriorityQueue(object):
	def __init__(self):
		self.queue = []

	def __str__(self):
		return ' '.join([str(i.priority) for i in self.queue])

	def isEmpty(self):
		return len(self.queue) == 0

	def insert(self, data):
		self.queue.append(data)

	def delete(self):
		max_val = 0
		for i in range(len(self.queue)):
			if self.queue[i].priority > self.queue[max_val].priority:
				max_val = i
		item = self.queue[max_val]
		del self.queue[max_val]
		return item.priority
        
    def printPatientQueue(self):
        print('->'.join([str(x.name) for x in self.items]))
```

```python
myQueue = PriorityQueue()
myQueue.insert(p1)
myQueue.insert(p2)
myQueue.insert(p3)
myQueue.insert(p4)
print(myQueue)		 
while not myQueue.isEmpty():
	print(myQueue.delete()) 
```
