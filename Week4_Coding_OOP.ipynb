#### Creating a class

```python
# Use the class statement to create a class
class Employee:
    # The init is also called a constructor, ie, for constructing a class instance
    # self denotes the newly created Employee instance. 
    def __init__(self, employee_type, salary):
        self.type = employee_type
        self.salary = salary    
```

```python
# This looks like a function but it is not. It creates a class instance. 
# Using the constructor prompts Python to run the __init__ function implicitly.
Peter = Employee('Long-term', 100000)
Tom = Employee('Volunteer',0)
```

```python
type(Peter)
```

```python
# Use object.attribute to access their attribute
Peter.salary
```

```python
# Accessing object attribute is different from using class methods
Peter.salary()
```

```python
# Define class specific methods 
class Employee:
    def __init__(self, employee_type, salary, year):
        self.type = employee_type
        self.salary = salary 
        self.year = year
    def change_type(self, new_type):
        self.type = new_type
    def add_year(self):
        self.year += 1
```

```python
# Use object.method() to call class specific methods on the instance
Peter = Employee('Long-term', 100000, 3)
Peter.add_year()
Peter.year
```

```python
# You can also directly modify the instance's attributes
Peter.year += 1
Peter.year
```

```python
# The methods are only accessible to instances within the class
a = 'Good'
a.change_type('Volunteer')
a
```

```python
# class data vs. instance data
class Employee:
    Employer = "UWM"
    def __init__(self, employee_type, salary):
        self.type = employee_type
        self.salary = salary    
```

```python
Amy = Employee("Full-time", 100)
Bob = Employee("Part-time", 40)
print(Amy.Employer, Bob.Employer) # class data/attribute is accessed by all class instances
print(Amy.type, Bob.type) # instance data/attribute is only accesed by the instance itself
```

### Inheriting from a super class

```python
class Employee:
    def __init__(self, employee_type, salary, year):
        self.type = employee_type
        self.salary = salary 
        self.year = year
    def change_type(self, new_type):
        self.type = new_type
    def add_year(self):
        self.year += 1
```

```python
class Student_Employee(Employee):
    # Note that this overrides the init function of its superclass
    # super().enables inheritance of some parts
    def __init__(self, employee_type, salary, year, year_graduate):
        super().__init__(employee_type, salary, year)
        self.year_graduate = year_graduate
    def is_graduate(self):
        if self.year_graduate <= 2025:
            return True
        else:
            return False
```

```python
Emily = Student_Employee('Intern', 5000, 0, 2022)
Emily.is_graduate()
```

```python
Emily.add_year()
Emily.year
```

```python
type(Emily)
```

#### Multiple inheritance

```python
class Person:
    def __init__(self, name):
        self.name = name
```

```python
# Student_Employee can inherit both from Employee and Person
class Student_Employee(Person, Employee):
    # Note that this overrides the init function of its superclass
    # super().enables inheritance of some parts
    def __init__(self, name, employee_type, salary, year, year_graduate):
        Employee.__init__(self, employee_type, salary, year)
        Person.__init__(self, name)
        self.year_graduate = year_graduate
    def is_graduate(self):
        if self.year_graduate <= 2025:
            return True
        else:
            return False
```

```python
Emily = Student_Employee("Emily", 'Intern', 5000, 0, 2022)
Emily.is_graduate()
```

```python
print(Emily.name)
print(Emily.type)
print(Emily.salary)
print(Emily.year)
print(Emily.year_graduate)
```

```python
Emily.add_year()
print(Emily.year)
```

```python
type(Emily)
```

```python
# method resolution order (MRO) can help you trace the inheritance of classes
print(Student_Employee.mro())
```

#### Customizing methods for subclasses for code reuse

```python
# Coding a method to get the last name of an employee
class Person:
    # self represents the newly created instance
    # it doesn't have to be named as self, but this is a convention
    def __init__(self, name, job=None, pay=0):
        self.name = name
        self.job = job
        self.pay = pay
    def get_lastname(self):
        return self.name.split()[-1]
    def giveRaise(self,percent):
        self.pay = int(self.pay * (1 + percent))
```

```python
# Inheritance
# Define a subclass Manager that shares similarities with the superclass
# But assume Managers get salary raise in a different way
class Manager(Person): 
    def giveRaise(self, percent, bonus=0.1):
        self.pay = int(self.pay * (1 + percent + bonus))
```

```python
carl = Manager('Carl Doe', pay=100000)
carl.giveRaise(0.2)
carl.pay
```

```python
# Instead of copying the giveRaise function from the super class, we can call the original function directly
# Code reuse and more efficient
class Manager(Person): 
    def giveRaise(self, percent, bonus=0.1):
        Person.giveRaise(self, percent + bonus)
```

```python
carl = Manager('Carl Doe', pay=100000)
carl.giveRaise(0.2)
carl.pay
```

### OOP Operator Overloading

Operator overloading allows you to define the behavior of standard operators (like +, -, *, print) for objects of your custom classes. 

By overloading operators, you can make them work in a way that’s specific to your class, enabling the use of operators in a more intuitive or domain-specific manner.


```python
# Coding a method to get the last name of an employee
class Person:
    # self represents the newly created instance
    # it doesn't have to be named as self, but this is a convention
    def __init__(self, name, job=None, pay=0):
        self.name = name
        self.job = job
        self.pay = pay
    def get_lastname(self):
        return self.name.split()[-1]
    def get_raise(self,percent):
        self.pay = int(self.pay * (1 + percent))
```

```python
amy = Person('Amy Doe')
bob = Person('Bob Doe',job='Engineer',pay=100000)
amy.get_lastname()
```

```python
# Printing an instance of the class is not very intuitive. 
print(bob)
```

```python
# Operator overloading
class Person:
    # self represents the newly created instance
    # it doesn't have to be named as self, but this is a convention
    def __init__(self, name, job=None, pay=0):
        self.name = name
        self.job = job
        self.pay = pay
    def get_lastname(self):
        return self.name.split()[-1]
    def giveRaise(self,percent):
        self.pay = int(self.pay * (1 + percent))
    # Called by the repr() built-in function to compute the “official” string representation of an object. 
    def __repr__(self):
        return '[Person: %s, %s]' % (self.name, self.pay)
```

```python
amy = Person('Amy Doe')
bob = Person('Bob Doe',job='Engineer',pay=100000)
print(bob)
```

1. Create a class called Point that has two attributes: x and y, denoting the x-axis value and y-axis value on a Cartesian coordinate. 

```python
class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y
```

2. Create an instance to represent a point at (0,0), the center of the coordiante. 

```python
p1 = Point(0,0)
p2 = Point(1,1)
p3 = Point(1,5)
```

3. Create a method calculated the distance between two points.

```python
import math
class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y
    def get_distance(self, p):
        d = math.sqrt((self.x - p.x)**2 + (self.y - p.y)**2 )
        return d
```

```python
p1 = Point(0,0)
p2 = Point(1,1)
p3 = Point(1,5)
p1.get_distance(p2)
```
