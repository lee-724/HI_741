### Queue

Queue is First in First out (FIFO).

Enqueue: It adds an item to the queue. If the queue is full, then it is said to be an Overflow condition.

Dequeue: It removes an item from the queue. The items are popped in the same order in which they are pushed. If the queue is empty, then it is said to be an Underflow condition.

Front: It gives the front item from the queue.

Rear: It gives the last item from the queue.

![title](Data_Queue.svg.png)

##### Using lists in Python to implement a queue

```python
class Queue:
    def __init__(self):
        self.items = []

    def isEmpty(self):
        return len(self.items) == 0

    def enqueue(self, item):
        self.items.append(item)

    def dequeue(self):
        return self.items.pop(0)

    def size(self):
        return len(self.items)
    
    def get_front(self):
        return self.items[0]

    def get_rear(self):
        return self.items[-1]
    
    def printQueue(self):
        print('->'.join([str(x) for x in self.items]))

    def printPatientQueue(self):
        print('->'.join([str(x.name) for x in self.items]))
```

```python
queue = Queue()
queue.enqueue(0)
queue.enqueue(1)
queue.enqueue(2)
print(queue.items)
queue.printQueue()
print("Front:", queue.get_front())
print("Rear:", queue.get_rear())
queue.dequeue()
queue.printQueue()
queue.dequeue()
queue.printQueue()
```

#### Using Python's built-in deque module (pronounced as: deck)

Deques are a generalization of stacks and queues (the name is pronounced “deck” and is short for “double-ended queue”). 

Deques support thread-safe, memory efficient appends and pops from either side of the deque with approximately the same O(1) performance in either direction.


```python
from collections import deque
  
q = deque()
  
q.append('a') # add a new entry to the right side
q.append('b')
q.append('c')
print(q)
q.appendleft('e')
  
print(q)
  
print(q.popleft())
print(q)
print(q.pop())
print(q)

q.count('a')
```
