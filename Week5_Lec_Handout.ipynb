### Overloading the index method by defining your own __getitem__

```python
class User:
    def __init__(self, username):
        self.username = username
        self.friends = []
    def add_friend(self, friend_user):
        self.friends.append(friend_user)
    def __getitem__(self, index):
        return self.friends[index]
```

```python
user1 = User("Tom")
user2 = User("Abby")
user1.add_friend(user2)
user1[0].username
```

### Implementing using different design patterns

#### Inheritance: Is-a relationship

```python
class Employee:
    def __init__(self, name, salary=0):
        self.name = name
        self.salary = salary
    def giveRaise(self, percent):
        self.salary = self.salary + (self.salary * percent)
    def work(self):
        print(self.name, "does stuff")
    def __repr__(self):
        return "<Employee: name=%s, salary=%s>" % (self.name, self.salary)
```

```python
class Chef(Employee):
    def __init__(self, name):
        Employee.__init__(self, name, 50000)
    def work(self):
        print(self.name, "makes food")
```

```python
class Server(Employee):
    def __init__(self, name):
        Employee.__init__(self, name, 40000)
    def work(self):
        print(self.name, "interacts with customer")
```

```python
chef = Chef("Amy")
chef.work()
```

```python
class Employee:
    def __init__(self, name, salary=0):
        self.name = name
        self.salary = salary
    def giveRaise(self, percent):
        self.salary = self.salary + (self.salary * percent)
        print(self.name, "got raise of ", percent, "%")
    def work(self):
        print(self.name, "does stuff")
    def __repr__(self):
        return "<Employee: name=%s, salary=%s>" % (self.name, self.salary)
```

```python
emp = Employee("Brian")
emp.giveRaise(5)
```

#### Aggregation: Has-a relationship

```python
class Customer:
    def __init__(self, name):
        self.name = name
    def order(self, server):
        print (self.name, "orders from", server.name)
    def pay(self, server):
        print (self.name, "pays to", server.name)
```

```python
class Oven:
    def bake(self):
        print("oven bakes")
```

```python
class PizzaShop:
    def __init__(self):
        # When you create a pizzashop, it automatically creates a chef and an oven. 
        self.server = Server("Dylan")
        self.chef = Chef("Amy")
        self.oven = Oven()
    def order(self, name):
        customer = Customer(name)
        customer.order(self.server)
        self.chef.work()
        self.oven.bake()
        customer.pay(self.server)
```

```python
scene = PizzaShop()
scene.order("Tim")
scene.order("Abby")
```
