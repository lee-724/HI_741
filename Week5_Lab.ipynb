### Interacting with files

#### Reading from a file

```python
# afile = open(file, mode)
# afile.method()
# mode: r for input, w for output, a for appending
```

```python
# Reading from a file
myfile = open('./test.txt', mode="r")
myfile.readline()
```

```python
myfile.readline()
```

```python
myfile.readline() # reaches the end of the file
```

```python
for line in open('./test.txt', "r"):
    print(line)
```

```python
l = []
with open('./test.txt', "r") as f:
    for line in f:
        print(line)
        l.append(line)
```

```python
l
```

#### Writing to a file

```python
# Writing to a file
myfile = open('myfile.txt', 'w')
myfile.write('Hello world\n')
```

```python
myfile.close()
```

```python
open('myfile.txt').read()
```

### Iterate through a directory

```python
import os
rootdir = './' # or you can change to any of your directory that you want to iterate over

for subdir, dirs, files in os.walk(rootdir):
    for file in files:
        # print(subdir)
        print(os.path.join(subdir, file))
```

```python
import os
rootdir = './'

for subdir, dirs, files in os.walk(rootdir):
    for file in files:
        if file.endswith('.txt'):
            print(os.path.join(subdir, file))
```
