### Linked List

From https://en.wikipedia.org/wiki/Linked_list

A linked list is a linear collection of data elements whose order is not given by their physical placement in memory.

Instead, each element points to the next. It is a data structure consisting of a collection of nodes which together represent a sequence. 

In its most basic form, each node contains: data, and a reference (in other words, a link) to the next node in the sequence. 
    
This structure allows for efficient insertion or removal of elements from any position in the sequence during iteration. 

![title](Singly-linked-list.svg.png)

```python
class Node:
    def __init__(self, value=None, next=None):
        self.value = value
        self.next = next
```

```python
class Node:
    def __init__(self, value=None, next=None):
        self.value = value
        self.next = next
        
class LinkedList:
    def __init__(self):  
        self.head = None
        
    def insert(self, value):
        newNode = Node(value)
        if(self.head):
            # head note empty = link list empty
            cur = self.head
            while(cur.next):
                cur = cur.next
            cur.next = newNode
        # Inserting to an empty list
        else:
            self.head = newNode
            
    def delete(self):
        cur = self.head
        prev = cur
        while (cur.next):
            prev = cur
            cur = cur.next
        prev.next = None
  
    def printLinkedList(self):
        cur = self.head
        while(cur):
            print(cur.value)
            cur = cur.next
```

```python
LL = LinkedList()
LL.insert(3)
LL.insert(4)
LL.insert(5)
LL.printLinkedList() # will give 3, 4, 5 in one row, not one line
LL.delete()
LL.printLinkedList()
```

```python
class LinkedList:
    def __init__(self):  
        self.head = None
        
    def insert(self, value):
        newNode = Node(value)
        if(self.head):
            cur = self.head
            while(cur.next):
                cur = cur.next
            cur.next = newNode
        # Inserting to an empty list
        else:
            self.head = newNode
            
    def delete(self):
        cur = self.head
        prev = cur
        while (cur.next):
            prev = cur
            cur = cur.next
        prev.next = None
  
    def printLinkedList(self):
        cur = self.head
        while(cur):
            print(cur.value)
            cur = cur.next
    def return_last(self):
        cur = self.head
        while (cur.next):
            cur = cur.next
        return cur
    def insert_after(self, prev_node, new_data):
        if prev_node is None:
            return
        new_node = Node(new_data)
        new_node.next = prev_node.next
        prev_node.next = new_node
```

```python
LL = LinkedList()
LL.insert(3)
LL.insert(4)
LL.insert(5)
LL.printLinkedList()
LL.insert_after(LL.return_last(), 6)
LL.printLinkedList()
```
