```python
# Sometimes it is unwanted to have a variable change after being passed to a function.
# We can avoid this by making a copy of that variable. 
def foo(a,b):
    a = 2
    b[0] = 's'
    print (b)

x = 1
l = [1, 2]
foo(x,l[:])
print(x,l)
```

```python
x = 88
def func():
    global x
    x = 99
func()
x
```

##### Arguments in function

```python
# Number of arguments should match in function call and function definition
def foo(a):
    print(a)
```

```python
foo()
```

```python
foo(1, 2)
```

```python
# There are multiple ways to pass arguments to a function. 
# Match by position
# May not be ideal when you have many parameters, hard to remember which value matches which parameter. 
def foo(a,b):
    a = 1
    b = 2
    print (a,b)
foo(0,1)
```

```python
# Match by keywords
# When you pass the values in by specifying the keywords, the orders/positions do not have to match. 
foo(b=1,a=0)
```

```python
# Default arguments
# Some arguments can be made as default and you don't need to pass in values for them unless you want to use other values
def foo(a, b=1, c=2):
    print(a,b,c)

foo(0)
foo(0,b=2)
foo(1,2,3)
```

```python
# Your function can also take any number of arguments
def f(*args):
    print(args)
f(1)
f(1,2,3)
f([0,1,2,3,4])
f()
```

```python
# You can also pass in values to be read as dictionaries
# Only works for keyword arguments not for positional arguments
def foo(**args):
    print(args)
foo()
foo(a=1,b=2,c=3)
```

### Functional programming in Python

Python supports more than one programming paradigms: procedural, object-oriented, and functional

Python has higher-order functions, which are functions that take one or more functions as arguments or returns a function.

#### Lambda

A lambda function is a small, anonymous function defined using the lambda keyword. 

Lambda functions are handy when you only need a short function temporarily. They are often used in combination with higher order functions such as map, filter, and reduce.


```python
# lambda arguments: expression
sum_val = lambda x, y: x + y
sum_val(1, 2)
```

#### Map

Mapping functions over iterables (e.g., lists)

```python
# Below is the traditional way of changing the value of each element in an iterable (list)
l = [0, 1, 2, 3]
new_l = []
for e in l:
    new_l.append(e * 2)
new_l
```

```python
# With map, it can be more readable and concise
# Map takes two arguments: a function and iterable. It returns a list with each element being applied that function
def multiply(x):
    return x * 2
list(map(multiply, l))
```

```python
# Or even better, lambda can be really handy here
list(map(lambda x: x * 2, l))
```

#### Filter

Filter can be used to select elements from iterables based on a function

```python
# Suppose we want to select elements that can be divided by 2
# The traditional way to do this is to write a loop
res = []
for x in range(-5, 5): 
            if x % 2 == 0:
                res.append(x)
res
```

```python
list(range(-5, 5))
```

```python
# Using filter and lambda can make your code much more concise and readable.
list(filter((lambda x: x % 2 == 0), range(-5, 5)))  # Select elements that can be divided by 2
```

#### Reduce

```python
from functools import reduce
# in Python 3 you need to manually import it
reduce((lambda x, y: x + y), [1, 2, 3, 4])
```

```python
reduce((lambda x, y: x * y), [1, 2, 3, 4])
```
