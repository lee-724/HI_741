```python
def basic_search(alist, item):
    for i, e in enumerate(alist):
        if item == e:
            return i
    return -1
```

```python
def generate_random_list(size):
    lst = []
    for _ in range(size):
        num = random.randint(0,100)
        lst.append(num)
    return lst
```

```python
lst = generate_random_list(2000)
basic_search(lst, 5)
```

```python
lst = generate_random_list(20000)
basic_search(lst, 19)
```

```python
def binarySearch(lst, item):
    first = 0
    last = len(lst)-1
    found = False
    while first<=last and not found:
        midpoint = (first + last)//2
        if lst[midpoint] == item:
            found = True
            return midpoint
        else:
            if item < lst[midpoint]:
                last = midpoint-1
            else:
                first = midpoint+1
    return midpoint
```

```python
lst = sorted(generate_random_list(200000))
binarySearch(lst, 19)
```
